```python
### Named Entity Recognization 
sentences ="""The Jagannath Temple is an important Hindu temple dedicated to Jagannath, a form of Vishnu – one of the trinity of supreme divinity in Hinduism. Puri is in the state of Odisha, on the eastern coast of India.
King Indradyumna of Avanti has built the main temple of Jagannath at Puri.
The present temple was rebuilt from the tenth 10th century onwards, on the site of pre-existing temples in the compound but not the main Jagannatha temple, and begun by Anantavarman Chodaganga, 
the first king of the Eastern Ganga dynasty. many rumours are spread about the temple but there is no solid proof of it.The temple is one of the 108 Abhimana Kshethram of Vaishnavate tradition. """
```

```python
print(sentences)
```

```python
import nltk
```

```python
words = nltk.word_tokenize(sentences)
```

```python
nltk.pos_tag(words)
```

```python
elements = nltk.pos_tag(words)
```

```python
import nltk
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
nltk.ne_chunk(elements)
```
